# Decision Tree:

**Decision tree algorithm is one of the most versatile algorithms in machine learning which can perform both classification and
regression analysis. It is very powerful and works great with complex datasets. Apart from that, it is very easy to understand
and read. That makes it more popular to use. When coupled with ensemble techniques  which we will learn very soon- it
performs even better. As the name suggests, this algorithm works by dividing the whole dataset into a tree-like structure based
on some rules and conditions and then gives prediction based on those conditions. It is called rule-based tree.**

**And most commonly , Decision Tree is used for Classfication Model , It works Good.**

**Let's understand the approach to decision tree with a basic scenario. Suppose it's Friday night and you are not able to decide if you should go out or stay at home. Let the decision tree decide it for you.**




Although we may or may not use Decision tree for such decisions , this was a basic example to help you understand that how a Decision Tree makes a Decision. So how did it work? 

**It selects a root node based on a given condition , e.g our root node was chosen as time >10 PM.** 

**All in between nodes(i.e no , other conditions/features) are tree depth , and the end nodes which decides finally , are called leaf nodes(Go ,Don't go).**

**Then , the root node was split into child nodes based on the given condition. The righr child node(yes) in the above figure fulfilled the condition , so no more questions were aske.**

**The left child node didn't fulfill the condition , so again it was split based on a new condition.**

**The process continues till all the conditions are met or if you have any predefined the depth of your tree e.g the depth of our tree is 3 ,and it reached there  when all the conditions were exhausted.**

**Important Question:**

As time and money are features here , so as like them there can be many features so how we know which features should come first , as here time comes first , and further how we know , which feature should come next ? 

**All these questions will be covered in Concepts/Criteria  'Entropy and Gini Index'.**


# Tree Pruning:

**Basically , it is used to save the time and money by trimming/cutting the depth of tree , to take a decision faster.**


Tree pruning is the method of trimming down a full tree (obtained through the above process) to reduce the complexity and variance in the data. Just as we regularised linear regression, we can also regularise the decision tree model by adding a new
term.




Where, T is the subtree which is a subset of the full tree TO And a is the non-negative tuning parameter which penalises the
MSE(Mean Square Error) with an increase in tree length. By using cross-validation, such values of a and T are selected for which our model gives the lowest test error rate. This is how the decision tree regression model works.

Let's now see the working algorithm of doing classification using a decision tree. Greedy Algorithm As per Hands-on machine learning book greedy algorithm greedily searches for an optimum split at the top level, then repeats the process at each level. It does not check whether or not the split will lead to the lowest possible impurity several levels down. A greedy algorithm often produces a reasonably good solution,but itis not guaranteed to be the optimal solution.


**Post Pruning:**

Post-pruning, also known as backward pruning, is the process where the decision tree is generated first and then the non-
significant branches are removed. Cross-validation set of data is used to check the effect of pruning and tests whether
expanding a node will make an improvement or not. If any improvement is there then we continue by expanding that node else
if there is reduction in accuracy then the node not be expanded and should be converted in a leaf node.

**Pre Pruning:**


Pre-pruning, also known as forward pruning, stops the non-significant branches from generating. It uses a condition to decide
when should it terminate splitting of some of the branches prematurely as the tree is generated.

# **Classification Trees:**

Regression trees are used for quantitative data. In the case of qualitative data or categorical data, we use classification trees.
In regression trees, we split the nodes based on RSS criteria, but in classification, it is done using classification error rate, Gini impurity and entropy. Let's understand these terms in detail

# **Entropy**

Entropy is the measure of randomness in the data. In other words, it gives the impurity present in the dataset.

**Ques. : So how do we know which feature should come first , or which next in a Decision Tree?**

Basically , Entropy find the measure of information gain in features and label , like which features has more information about label.

**For example , let us consider  phone no. ,age , city , exp , technology are features , and salary is label , in any Dataset. so we will see which of the feature has more information about , now in these features , obviously exp has more information about salary , because more the exp , more will be salary , so 'exp' will come first and called root node in this example. And on the basis of exp , there will be more conditions and a Decision tree will start to build.**

Similarly , the next feature which has more information of salary will come to the decision tree , then there will be conditions also and decision tree wil going on.


# Gini Impurity/Indexing:

**Work role is similar to Entropy , but the difference is gini will find impurity in columns/features.**

**In entropy , we discussed that Higher information gain feature will be priortized , But in the Gini Indexing , we will find 'Lowest Impurity'.**

**Suppose any feature has lowest impurity , it means it has higher information gain , so that feature can come first as root node in Decision Tree.**

**Gini Impurity values lies b/w 0 and 1 . 0 means no impurity ,and 1 means random distribution. The node for which the Gini Impurity is least is selected as the root node to split.**

# Different Algorithms For Decision Tree:

**1.ID3(Iterative Dichotomiser): It is one of the algorithms used to construct Decision Tree for Classification. It uses Information Gain as the Criteria for finding the root nodes and splitting them. It only accepts Categorical Attributes.**

**C4.5 : It is an extension of ID3 algorithm , and Better than ID3 as it deals with both Continuous and Discrete values . It is also used for Classification purposes.**

**Classification and Regression Trees(CART): It is the most popular algorithm used for constructing Decision Trees. It uses Gini Impurity as as the default calculation for selecting root nodes,however one can use entropy for criteria as well. This algorithm works on Both Regression as well as Classification Problems. We will use this algorithm in our python Implementation.**

Entropy and Gini Impurity can be used reversibly . It doesn't effects the result much . Although Gini is much easier to compute than Entropy . Since Entropy has log term calculation.That's why CART algorithm uses Gini as the default algorithm.

**if we plot Gini vs Entropy graph , we can see there is not much difference b/w them:**

```python
# Let's define X and y:

X = data.drop(columns =['quality','Alcohol_content'])
y= data['quality']
```

```python
# Let's split the data for training and testing:

x_train , x_test , y_train , y_test = train_test_split(X, y , test_size = 0.25 ,random_state = 41)
```

```python
# Let's write one function and call as many times to check accuracy score of different models:

def metric_score(clf , x_train , x_test , y_train , y_test , train=True):
    if train:
        y_pred = clf.predict(x_train)

        print("\n==============Train Result==========")
        
        print(f"=======accuracy_score:{accuracy_score(y_train , y_pred)*100:.2f}%")
        
    elif train ==False:
        
        pred = clf.predict(x_test)
        
        print("\n===============Test Result===========")
        
        print(f"accuracy_score:{accuracy_score(y_test , pred)*100:.2f}%")
        
        print('\n \n Test Classification Report \n',classification_report(y_test ,pred,digits=2)) #Model confindence/accurcy.
```

```python
# Model Initiation:

clf = DecisionTreeClassifier()
clf.fit(x_train , y_train)  # for training
```

```python
# Call the function and pass dataset to check train and test score:

metric_score(clf,x_train,x_test , y_train , y_test ,train= True)  # This is for training score
metric_score(clf,x_train,x_test , y_train , y_test ,train= False)  # This is for testing score
```

**As training score is 100% , it means decision tree learned/understood everything bit by  bit , but in testing there is drop as it is just 60 % , it happens because in the test data set , we had less number of records/less data. But we didn't expected thus drop 100 to 60%. We will try to improve using hyperparameter tuning.**

 # Model Confidence /Accuracy:
 
 **Let's now try to tune some hyperparameters using the GridSearchCV Algorithm.**
 
 **GridSearchCV is a method used to tune our hyperparameters. We can pass different values of Hyperparameters as parameters for GridSearch . It does a exhaustive generation of combination of different parameters passed.**
 
 Using Cross validation score , GridSearch returns the combination of Hyperparameters for which the model is performing the best.
 
 

**What are Hyperparameters??**



**We can see above , The decision Tree classsifier algorithm takes all those parameters which are also known as Hyperparameters.**

**When we do hyperparameter tuning , we basically try to find those sets and values  of hyperparameters which will gives us a model with maximum accuracy.Let's go ahead and try to improve out model.**

```python
# We are tuning four important hyperparameters right now ,We are passing different values for both;

grid_param = {'criterion':['gini','entropy'],
             'max_depth': range(10,15),       #The maximum depth of Tree
             'min_samples_leaf': range(2,6),   # The min. no. of samples required to be at leaf node.
             'min_samples_split': range(3,8),  # The min. no. of samples required to split an interval node.
             'max_leaf_nodes': range(5,10)}    # Best nodes are defined as the relatve reduction in impurity. If none then unlimited no. of leaf nodes.
```

```python
grid_search = GridSearchCV(estimator = clf , param_grid = grid_param ,cv= 5 , n_jobs = -1)

# estimator means which model we are using.
# param_grid will take the parameters.
# cv is cross validation , don't increase cv , so time and money also increases.
# n_jobs = -1 , if system taking lot of time , then use this , it will improve little bit of performance.
```

```python
grid_search.fit(x_train , y_train)
```

```python
best_parameters = grid_search.best_params_
best_parameters
```

```python
# Initiate DecisionTreeClassifier with new parameters and train:

clf = DecisionTreeClassifier(criterion = 'entropy',min_samples_split = 3 ,max_leaf_nodes= 7, max_depth = 10 , min_samples_leaf =2)

# Train the  model:

clf.fit(x_train , y_train)
```

```python
# AGAIN , Call the function and pass dataset to check train and test score:

metric_score(clf,x_train,x_test , y_train , y_test ,train= True)  # This is for training score
metric_score(clf,x_train,x_test , y_train , y_test ,train= False)  # This is for testing score
```
